```python
def corey_Krw(sw,swc,sor,nw):
    sw_norm = (sw-swc)/(1-swc-sor)
    sw_norm = max(0,min(1,sw_norm))

    krw = sw_norm**nw
    return krw 
```

```python
krw = corey_Krw(0.4,0.2,0.25,2.5)
```

```python
#f(x) = x3 - sinx
#f'(x) = 3x^2 -cosx
```

```python
def fx(x):
    y = x**3 - np.sin(x)

    return y
```

```python
def fdx(x):
    ydx = 3*(x**2) - np.cos(x)

    return ydx 
```

```python
x_0 = float(input("Hey, enter the initial guess for root of equation x^3-sinx: "))

x= x_0 
count = 0

print(f"The value of x for {count} iteration is {x}")
while abs(fx(x))>0.0000000001:
    print("=================")
    y = fx(x)
    ydx = fdx(x)
    x = x -y/ydx
    print(f"The value of x for {count} iteration is {x}")
    count+=1

print(f"The root is {x}")
```

```python
def root(x_0):

    """
    Root is a function that finds the root of equation x^3 - sin(x).

    input parameters:
    x_0 = initial guess for the root

    returns:
    x = actual root of the function
    count = after how many iterations we reached to the root

    
    """

    x= x_0 
    count = 0

    print(f"The value of x for {count} iteration is {x}")
    while abs(fx(x))>0.0000000001:
        print("=================")
        y = fx(x)
        ydx = fdx(x)
        x = x -y/ydx
        print(f"The value of x for {count} iteration is {x}")
        count+=1

    print(f"The root is {x}")

    return x,count
        
```

```python
root(434343)
```
